```python
import pandas as pd

df = pd.read_csv("/employees.csv")
df.head()
```

```python
 # 3. Clean data
df = df.drop_duplicates()
df["hire_date"] = pd.to_datetime(df["hire_date"])
df["salary"] = pd.to_numeric(df["salary"], errors="coerce")
```

```python
# 4. Data quality checks
quality_report = {
    "missing_employee_ids": df["employee_id"].isnull().sum(),
    "duplicate_records": df.duplicated().sum(),
    "missing_departments": df["department"].isnull().sum(),
    "invalid_salaries": (df["salary"] <= 0).sum()
}
```

```python
# 5. Create HR metrics
summary = df.groupby("department").agg(
    headcount=("employee_id", "count"),
    avg_salary=("salary", "mean"),
    attrition_count=("attrition", "sum")
).reset_index()

summary["attrition_rate"] = summary["attrition_count"] / summary["headcount"]
```

```python
import os

os.makedirs("sample_outputs", exist_ok=True)
```

```python
# 6. Export files
df.to_csv("sample_outputs/clean_employee_data.csv", index=False)
summary.to_csv("sample_outputs/hr_summary_metrics.csv", index=False)
pd.DataFrame([quality_report]).to_csv("sample_outputs/data_quality_report.csv", index=False)
```
